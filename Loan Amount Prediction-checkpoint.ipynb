## Features Engineering 

```python
df.columns
```

```python
### Remove unnecessary columns

df=df.drop(columns=['customer_id','application_date','vintage_year','vintage_month'])
```

```python
### Making features that capture repayment capacity

df['income_to_emi']=df['monthly_income']/df['emi']
```

```python
### Creating features that tell us the loan affordibility

df['loan_to_income']=df['loan_amount']/df['annual_income']
```

```python
### Making feature that show credit pressure

df['credit_pressure']=(df['credit_inquiries_12m'] * (df['revolving_utilization_origination'] / 100))   
```

```python
df.sample(5)
```

```python
df.shape
```

```python
### split data into x,y 

x=df.drop('loan_amount',axis=1)
y=df['loan_amount']
```

```python
### Further split data into train test split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
### Print spliting distribution

print('x_train shape:',x_train.shape)
print('x_test shape:',x_test.shape)
print('y_train shape:',y_train.shape)
print('y_test shape:',y_test.shape)
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
### Encode and scale the categorical data into onehotencoder and Lableencoder Using Coli


nums = x_train.select_dtypes(include='number').columns.tolist()
cat=['gender','employment_status']
cats=['deterioration_pattern']

ct=ColumnTransformer(
    transformers=[
        ('ohe',OneHotEncoder(drop='first',sparse_output=False,handle_unknown='ignore'),cat),
        ('ord', OrdinalEncoder(), cats),
        ('scale',StandardScaler(),nums)
        
    ],remainder='passthrough'
)

x_train=ct.fit_transform(x_train)
x_test=ct.transform(x_test)
```

```python
x_train
```

```python
x_test
```

## Train & Evaluate the model with tree and non-tree based algorithms

```python
from sklearn.metrics import mean_squared_error
```

```python
models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'DecisionTreeRegressor':DecisionTreeRegressor(),
    'RandomForestRegressor':RandomForestRegressor(),
    'GradientBoostingRegressor':GradientBoostingRegressor(),
    'XGB':XGBRegressor(),
    'Extratree':ExtraTreeRegressor(),
    'Lighgbm':lgb.LGBMRegressor(),
    'Kneighbors':KNeighborsRegressor()
}

results = []

for name, model in models.items():
    
    model.fit(x_train, y_train)
    
    
    y_pred = model.predict(x_test)
    
    
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    

    results.append({
        'Model': name,
        'R2_Score': r2,
        'MAE': mae,
        'RMSE': rmse
    })


results_df = pd.DataFrame(results).sort_values(by='R2_Score', ascending=False)
results_df
```

Tree-based ensemble models performed best due to nonlinear relationships.
Linear models underperformed, indicating strong feature interactions.
RandomForest showed near-perfect accuracy,I selected GradientBoosting as the final model due to better generalization.

## Pick the final model 'GradientBoosting'

```python
### Train & Evaluate model with Gradientboosting

gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)
y_predd=gb.predict(x_test)

print('R2 Score :',r2_score(y_test,y_predd))
print('MAE Score :',mean_absolute_error(y_test,y_predd))
print('RMSE Score :',root_mean_squared_error(y_test,y_predd))    
```

```python
### Check model generalize well or just variance bias

y_train_pred = gb.predict(x_train)
y_test_pred = gb.predict(x_test)

print("Train R:", r2_score(y_train, y_train_pred))
print("Test R:", r2_score(y_test, y_test_pred))
```

GradientBoosting model generalizes well, with nearly identical train and test R scores, indicating low variance and low bias. The small performance gap confirms absence of overfitting and suggests robust learning of underlying patterns.
